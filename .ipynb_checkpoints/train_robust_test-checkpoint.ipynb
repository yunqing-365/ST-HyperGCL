```python


import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import roc_auc_score, accuracy_score
import warnings
warnings.filterwarnings("ignore")

from models.st_hypergcl import UltimateRiskModel, contrastive_loss

print("加载时序动态张量数据")

# 1. 加载切片数据
X = torch.load("data/node_features_X.pt")
Y = torch.load("data/risk_labels_Y.pt")
edge_index_list = torch.load("data/dynamic_edge_indices.pt")

# 标准化初始特征
X_mean = X.mean(dim=0)
X_std = X.std(dim=0) + 1e-5
X_normalized = (X - X_mean) / X_std

num_companies = X.shape[0]
T = len(edge_index_list)

# 2. 预处理：生成邻接矩阵 A
adj_list = []
for t in range(T):
    edge_index_t = edge_index_list[t]
    if edge_index_t.shape[1] == 0:
        adj_list.append(torch.eye(num_companies))
        continue
        
    num_inventors_t = edge_index_t[0].max().item() + 1
    H = torch.zeros((num_companies, num_inventors_t))
    H[edge_index_t[1], edge_index_t[0]] = 1.0 
    
    A = torch.matmul(H, H.t())
    A.fill_diagonal_(1.0)
    deg = A.sum(dim=1, keepdim=True)
    A_norm = A / deg
    adj_list.append(A_norm)

# 3. 实例化模型与优化器
model = UltimateRiskModel(input_dim=X.shape[1])
optimizer = optim.Adam(model.parameters(), lr=0.005, weight_decay=1e-4)
pos_weight = torch.tensor([5.0]) 
criterion_bce = nn.BCEWithLogitsLoss(pos_weight=pos_weight)

# 4. 开始训练
epochs = 200
alpha = 0.5 # 对比学习权重 (改成0进行无去噪的消融实验)

# ... 前面的加载数据代码都不变 ...

print("\n🚀 点火！开始【10% 恶性高斯噪声】抗打击测试...")

for epoch in range(epochs):
    model.train()
    optimizer.zero_grad()
    
    # ==========================================
    # 👿 核心摧残：往完美特征里注入 10% 的高斯噪声
    # 模拟真实世界里被粉饰、造假的脏数据
    # ==========================================
    noise_level = 0.10
    noise = torch.randn_like(X_normalized) * noise_level
    noisy_X = X_normalized + noise
    
    # 前向传播：注意这里喂给模型的是被污染的 noisy_X，而不是原始的 X_normalized！
    logits, view1, view2 = model(noisy_X, adj_list)
    
    # 计算双重 Loss
    loss_bce = criterion_bce(logits, Y)
    loss_gcl = contrastive_loss(view1, view2)
    
    # 终极 Loss
    loss = loss_bce + alpha * loss_gcl
    
    loss.backward()
    optimizer.step()
    
    if (epoch + 1) % 20 == 0:
        model.eval()
        with torch.no_grad():
            # ⚠️ 测试的时候，我们依然用被污染的数据来刁难模型！
            test_noise = torch.randn_like(X_normalized) * noise_level
            test_noisy_X = X_normalized + test_noise
            
            preds_prob = torch.sigmoid(model(test_noisy_X, adj_list)[0]).numpy()
            true_y = Y.numpy()
            
            preds_label = (preds_prob > 0.5).astype(float)
            acc = accuracy_score(true_y, preds_label)
            try:
                auc = roc_auc_score(true_y, preds_prob)
            except ValueError:
                auc = 0.5
            
            print(f"Epoch [{epoch+1:03d}/{epochs}] | BCE: {loss_bce.item():.4f} | GCL: {loss_gcl.item():.4f} | 🛡️ 抗噪 AUC: {auc:.3f}")

print("\n🏆 暴力摧残结束！请查看最终抗噪 AUC！")
```
