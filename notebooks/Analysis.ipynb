```python
import pandas as pd
import os

# ==========================================
# 阶段 1：CSMAR 真实数据自动化探查与字段提取
# ==========================================

# 配置文件名
files = {
    "1. 研发联盟 (超图拓扑)": "CA_EnterpriseRDAlliance.csv",
    "2. 财务特征 (节点初始X)": "FS_Comins(Merge Query).csv",
    "3. 违规标签 (预测目标Y)": "STK_Violation_Main.csv",
    "4. 高管网络 (核心节点)": "TMTLI_FigureInfo.csv"
}

def peek_csv_data(file_name, display_name):
    """
    安全读取 CSMAR 数据，自动处理中文编码，并打印表头和第一行样本
    """
    if not os.path.exists(file_name):
        print(f"❌ 找不到文件: {file_name}")
        return
    
    # CSMAR 数据常见两种编码，尝试自动解析
    try:
        df = pd.read_csv(file_name, encoding='utf-8', nrows=5)
    except UnicodeDecodeError:
        df = pd.read_csv(file_name, encoding='gbk', nrows=5)
    except Exception as e:
        print(f"读取 {file_name} 发生未知错误: {e}")
        return

    print(f"\n{'='*50}")
    print(f"📊 {display_name} | 文件: {file_name}")
    print(f"{'='*50}")
    
    # 打印所有的列名，方便我们后续写代码进行精确映射
    columns = df.columns.tolist()
    print(f"📌 包含的列名 (共 {len(columns)} 列):")
    print(columns)
    
    print("\n🔍 第一行真实数据样本:")
    # 截取前几个最重要的字段展示即可，防止屏幕打印太满
    sample_dict = df.iloc[0].to_dict()
    for k, v in list(sample_dict.items())[:8]: 
        print(f"   - {k}: {v}")

# 批量执行探查
for name, file_path in files.items():
    peek_csv_data(file_path, name)
```
